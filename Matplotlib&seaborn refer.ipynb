```python
import numpy as np

# Creating the arrays
first_array = np.array([10, 12])
second_array = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Iterating through the arrays and printing the elements
for i in range(len(first_array)):
    print("First array element:", first_array[i])
    for j in range(len(second_array[i])):
        for k in range(len(second_array[i][j])):
            print("Second array element:", second_array[i][j][k])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean = 0
sigma = 1

data = np.random.normal(mean, sigma, 1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

pdf = stats.norm.pdf(np.linspace(min(data), max(data), 100), mean, sigma)
plt.plot(np.linspace(min(data), max(data), 100), pdf,  linewidth=2)

plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success
num_experiments = 1000

# Simulate the experiments
results = np.random.binomial(n, p, num_experiments)

# Calculate the PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plot the PMF
plt.hist(results, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='g', edgecolor='black')
plt.plot(x, pmf, 'o', color='black')
plt.title('Probability Mass Function (PMF) of Binomial Distribution')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
```
